```python
import matplotlib.pyplot as plt
import numpy as np
```

# 11.1

```python
#1)
month = [7, 8, 9, 10, 11, 12]
user = [456, 492, 578, 599, 670, 854]

plt.plot(month, user, color='blue', linestyle='solid')

plt.title("Daehan company speeda net new customers")

plt.xlabel("Month")
plt.ylabel("User")
plt.show()
```

```python
#2)
fig, ax = plt.subplots(1, 4, figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.bar(month, user)
plt.title('Bar chart')

plt.subplot(1, 4, 2)
plt.plot(month, user, linestyle='solid')
plt.title('Line chart')

plt.subplot(1, 4, 3)
plt.scatter(month, user, marker = '^')
plt.title('Scatter chart')

plt.subplot(1, 4, 4)
plt.barh(month, user)
plt.title('Horizontal Bar chart')

plt.show()
```

# 11.2

```python
a = np.arange(0, (2*np.pi)*6, 0.1) #1의 간격으로 하니 뻣뻣하게 그래프가 나와서 부드럽게 만들기 위해 0.1 간격으로 생성하다.
x = np.sin(a)
y = np.cos(a)

plt.plot(a, x*a, color='red', linestyle='solid')
plt.plot(a, y*20, color='blue', linestyle='--')

plt.show()
```

# 11.3

```python
a = np.random.randint(0, 50, size=60) #랜덤으로 0부터 50까지의 숫자 60개를 생성한다.
ndarray = a.reshape(30, -1) #30:2의 2차원 배열로 만들어줌

for i in range(0, 30) : 
    plt.scatter(ndarray[i][0], ndarray[i][1], color = 'blue')

plt.show()
```

# 11.4

```python
fig, ax = plt.subplots(1,3, figsize = (15, 5))

#N(a, b**2)[평균 a, 표준편차b]에서 정규분포를 갖는 난수를 생성하기 위해서는
# b*np.random.randn(...) + a 의 형태를 사용한다.

#x값과 y값이 각각 평균이 25이고 표준편차가 6
x1 = 6*np.random.randn(1000) + 25
y1 = 6*np.random.randn(1000) + 25

plt.subplot(1,3,1)
plt.scatter(x1, y1, color = 'blue')
plt.axis([0,50,0,50])

#x값의 평균이 25, 표준편차 6/ y값의 평균은 25, 표준편차 3
x2 = 6*np.random.randn(1000) + 25
y2 = 3*np.random.randn(1000) + 25

plt.subplot(1,3, 2)
plt.scatter(x2, y2, color = 'red')
plt.axis([0,50,0,50])

#x값의 평균이 25, 표준편차 6/ y값은 x값에 표준정규분포 N(1,0) 난수를 더한 값
x3 = 6*np.random.randn(1000) + 25
y3 = x3 + 1 

plt.subplot(1,3,3)
plt.scatter(x3, y3, color='green')
plt.axis([0,50,0,50])

plt.show()
```

# 11.5

```python
#닥스훈트
d_length = [77, 78, 85, 83, 73, 77, 73, 80]
d_height = [25, 28, 29, 30, 21, 22, 17, 35]

#사모예드
s_length = [75, 77, 86, 86, 79, 83, 83, 88]
s_height = [56, 57, 50, 53, 60, 53, 49, 61]

#말티즈
m_length = [34, 38, 38, 41, 30, 37, 41, 35]
m_height = [22, 25, 19, 30, 21, 24, 28, 18]
```

```python
#1)
fig, ax = plt.subplots(1,3, figsize = (20, 5))

plt.subplot(1, 3, 1)
plt.scatter(d_length, d_height, color = 'red', marker = 'o')
plt.title('Dachshund size')
plt.xlabel('Length')
plt.ylabel('Height')

plt.subplot(1, 3, 2)
plt.scatter(s_length, s_height, color = 'blue', marker = '^')
plt.title('Samoyed size')
plt.xlabel('Length')
plt.ylabel('Height')

plt.subplot(1,3, 3)
plt.scatter(m_length, m_height, color = 'green', marker = 's')
plt.title('Maltese size')
plt.xlabel('Length')
plt.ylabel('Height')

plt.show()
```

```python
#2)
plt.scatter(d_length, d_height, color = 'red', marker = 'o', label='Dachshund')
plt.scatter(s_length, s_height, color = 'blue', marker = '^', label='Samoyed')
plt.scatter(m_length, m_height, color = 'green', marker = 's', label='Maltese')

plt.xlabel("Length")
plt.ylabel("Height")
plt.title("Dog size")
plt.legend()
plt.show()
```
